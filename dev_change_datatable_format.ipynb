```python
from project_huishoudboekje.config import GeneralSettings

import numpy as np
import pandas as pd
import uuid
```

```python
df = pd.read_excel(GeneralSettings.project_path / 'data/processed/transactions.xlsx')
```

```python
df
```

```python
df['TRANS_ID'] = df['TRANS_ID'].apply(lambda x: uuid.uuid4())
```

```python
df = df.drop(['Unnamed: 0.1', 'id', 'Unnamed: 0', 'SUB_ITEM'], axis=1)
```

```python
df = df.rename(columns={'SUB_CATEGORY': 'CATEGORY'}).assign(SOURCE='Rabobank')
```

```python
df = df.reindex(columns=['TRANS_ID', 'DATE', 'SOURCE', 'TRANSACTION_TYPE', 'FINANCIAL_TYPE', 'PARTY',
       'AMOUNT', 'CATEGORY', 'ANALYSE_IND'])
```

```python
df.to_excel(GeneralSettings.project_path / 'data/processed/transactions.xlsx', index=False)
```

```python
df_categories = pd.read_excel(GeneralSettings.project_path / 'data/categories.xlsx')
```

```python
df = df.merge(df_categories, how='left', left_on='CATEGORY', right_on='CATEGORY')
```
